```
X = [1,5,1.5,8,1,9]
y = [2,8,1.8,8,0.6,11]
X = np.array([
             [1,2],
             [5,8],
             [1.5,1.8],
             [8,8],
             [1,0.6],
             [9,11]])     

print(X)
y = [0,1,0,1,0,1]
```

```
clf = svm.SVC(C=1.0, kernel='linear')
clf.fit(X,y)
```

```
clf.predict([[0.56, 0.78],[7,9]])
```

```
w = clf.coef_[0]
print(w)
```

```
print(w[0], w[1])
```

```
a = -w[0]/w[1]
print(a)
```

```
clf.intercept_[0]
```

```
xx = np.linspace(0,12)
a*xx
yy = a*xx - (clf.intercept_[0]) / w[1]
```

```
print(a*xx)
print(type(a*xx))
print(yy)
print(type(yy))
```

```
plt.plot(xx, yy, )
```

```
h = plt.plot(xx, yy, 'k-', label="non weighted divide")
plt.show()
```
